# **PREPROCESSING**

```python
df['Name'] = df['Name'].str.replace('\d+', '')
```

```python
#REMOVE PUNCTUATION
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
df['clean_msg']= df['Name'].apply(lambda x:remove_punctuation(x))
df
```

```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
#REMOVE ESCAPE
def remove_escape(s):
  escapes = ''.join([chr(char) for char in range(1, 32)])
  translator = str.maketrans('', '', escapes)
  t = s.translate(translator)
  return t
df['clean_msg']= df['clean_msg'].apply(lambda x:remove_escape(x))
df.head()
```

```python
# TO LOWER CONVERSION
df['clean_msg']= df['clean_msg'].apply(lambda x: x.lower())
df.head()
```

```python
#WORD TOKENIZE
df['clean_msg']= df['clean_msg'].apply(lambda x: word_tokenize(x))
df
```

```python
all_stopwords=stopwords.words('english')
all_stopwords.append(' ')
all_stopwords
```

```python
def remove_stopwords(text):
  stopwordfree=[i for i in text if i not in all_stopwords]
  return stopwordfree
df['clean_msg']= df['clean_msg'].apply(lambda x: remove_stopwords(x))
df
```

```python
#STEMMING
porter_stemmer = PorterStemmer()
#defining a function for stemming
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
df['clean_msg']=df['clean_msg'].apply(lambda x: stemming(x))
df
```

```python
#LEMMATIZATION
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
#defining the function for lemmatization
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
df['clean_msg']=df['clean_msg'].apply(lambda x:lemmatizer(x))
df
```

```python
df['clean_msg']=df['clean_msg'].apply(lambda x:' '.join(x))
df
```

# **VECTORIZATION**

```python
corpus=df['clean_msg'].tolist()
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
vectors=vectorizer.fit_transform(corpus)
feature_names=vectorizer.get_feature_names()
print("Feature names ",feature_names)
matrix=vectors.todense()
ndf=pd.DataFrame(matrix,columns=feature_names)

print("la",vectors)
```

# **VECTOR FOR QUERTY TO SEARCH**

```python
from sklearn.metrics.pairwise import linear_kernel
q=['markets offer whether capture ancient ']
qv=vectorizer.transform(q)
print("check",qv)

cosine_similarities = linear_kernel(qv, vectors).flatten()
cosine_similarities
```

# **Cosine Similarity function for the calculation**

```python
related_docs_indices = cosine_similarities.argsort()[:-5:-1]
related_docs_indices
```
